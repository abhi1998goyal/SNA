```python
threshold_5 = np.percentile(list(degrees.values()), 95)
```

```python
first_5_nodes = [node for node, degree in degrees.items() if degree >= threshold_5]
G_top_5 = G.subgraph(first_5_nodes)
```

```python
plt.figure(figsize=(10, 10))
pos = nx.spring_layout(G_top_5)
nx.draw(G_top_5, pos, node_size=50, with_labels=False)
plt.title('Top 5% Nodes by Degree')
plt.show()
```
